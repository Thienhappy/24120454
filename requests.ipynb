```python
def read_text_file(file_path):
    """
    Đọc nội dung từ file .txt và trả về chuỗi văn bản.
    """
    if not os.path.exists(file_path):
        print(f"⚠️ Lỗi: Không tìm thấy file tại đường dẫn: {file_path}")
        return None
    
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            content = f.read()
            return content
    except Exception as e:
        print(f"⚠️ Lỗi khi đọc file: {e}")
        return None
```
